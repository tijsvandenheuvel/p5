```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_as_arr=[]
with open("data2.txt", "r") as file:
    for line in file:
        line_arr = line.split()
        if(len(line_arr)>8):
            data_as_arr.append([line_arr[1],line_arr[3],line_arr[5],line_arr[7],line_arr[9],line_arr[11]])
```

```python
data_as_arr[0]
```

```python
df = pd.DataFrame(data_as_arr) 
df.columns = ['gen', 'agents','food','speed','force','health']
```

```python
df.count()
```

## agents vs food

```python
x = df['gen'].astype(float).to_numpy()
y1 = df['agents'].astype(float).to_numpy()
y2 = df['food'].astype(float).to_numpy()

plt.plot(x,y1,label='agents')
plt.plot(x,y2,label='food')


z1 = np.polyfit(x, y1, 1)
p1 = np.poly1d(z1)
plt.plot(x,p1(x),"r--")

z2 = np.polyfit(x, y2, 1)
p2 = np.poly1d(z2)
plt.plot(x,p2(x),"r--")

plt.legend()
plt.show()
```

```python
# grow rate agents
print(z1[0])
# grow rate food
print(z2[0])
```

## speed

```python
x = df['gen'].astype(float).to_numpy()
y = df['speed'].astype(float).to_numpy()

plt.scatter(x,y)
#plt.plot(df['gen'],df['force'])

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")
#plt.gca().invert_yaxis()
plt.show()
```
