```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_as_arr=[]
with open("data2.txt", "r") as file:
    for line in file:
        line_arr = line.split()
        if(len(line_arr)>8):
            data_as_arr.append([line_arr[1],line_arr[3],line_arr[5],line_arr[7],line_arr[9]])
```

```python
data_as_arr[0]
```

```python
df = pd.DataFrame(data_as_arr) 
df.columns = ['gen', 'agents','food','speed','force']
```

```python
df.count()
```

```python
plt.plot(df['gen'],df['agents'])
plt.plot(df['gen'],df['food'])

#plt.xticks(np.arange(df.gen.count(), step=20))
#plt.yticks(np.arange(50, step=10))

plt.show()
```
